```python
import pandas as pd 
```

```python
df = pd.read_csv('../../Data/Pharmacy/D2.csv')
```

```python
df.head()
```

```python
df[['date', 'time']] = df['date'].str.split(' ', 1, expand=True)
```

```python
df.head()
```

```python
df = df.drop('date', axis=1)
```

```python
df.head()
```

```python
df = df.rename(columns={'datestring': 'date'})
```

```python
df.head()
```

```python
df[['expiry_date', 'expiry_time']] = df['expiry'].str.split(' ', 1, expand=True)
```

```python
df.head()
```

```python
df = df.drop('expiry', axis=1)
```

```python
df.head()
```

```python
df = df.drop('expiry_time', axis=1)
```

```python
df.head()
```

```python
df = df.rename(columns={'expiry_date': 'expiry'})
```

```python
df.head()
```

```python
looseqty_column = df.pop('date')
df['date'] = looseqty_column
```

```python
looseqty_column = df.pop('time')
df['time'] = looseqty_column
```

```python
looseqty_column = df.pop('itemname')
df['itemname'] = looseqty_column
```

```python
looseqty_column = df.pop('packunits')
df['packunits'] = looseqty_column
```

```python
looseqty_column = df.pop('expiry')
df['expiry'] = looseqty_column
```

```python
looseqty_column = df.pop('price')
df['price'] = looseqty_column
```

```python
looseqty_column = df.pop('looseqty')
df['looseqty'] = looseqty_column
```

```python
df.head()
```

```python
df.to_csv('../../Data/Pharmacy/D3.csv', index=False)
```
