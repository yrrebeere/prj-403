```python
import pandas as pd
```

```python
df = pd.read_csv('../../Data/Pharmacy/C3.csv')
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
```

```python
df.head()
```

```python
agg_functions = {
    'packunits': 'first',
    'expiry': 'first',
    'price': 'sum',
    'looseqty': 'sum'
}
```

```python
combined_df = df.groupby(['date', 'itemname'], as_index=False).agg(agg_functions)
```

```python
combined_df.head()
```

```python
combined_df['date'] = combined_df['date'].dt.strftime('%d/%m/%Y')
```

```python
combined_df.head()
```

```python
combined_df.to_csv('../../Data/Pharmacy/C4.csv', index=False)
```
